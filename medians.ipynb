```python
def find_median(df):
    df['Price'] = df['Price'].str.replace(',', '')
    df['Price'] = df['Price'].replace({r'\$':''}, regex = True).replace(",", "")
    cols_as_np = (df[df.columns[2]].to_numpy()).astype(int)
    return np.median(cols_as_np)
    
```

```python
cities_med = np.array([])
for i in range(len(cities)):
    print(cities[i] + ':', find_median(cities_df[i]))
    cities_med = np.append(cities_med, find_median(cities_df[i]))
```

```python
markdict = dict(zip(cities, cities_med))
marklist=list(markdict.items())
l=len(marklist)
for i in range(l-1):
    for j in range(i+1,l):
        if marklist[i][1]>marklist[j][1]:
            t=marklist[i]
            marklist[i]=marklist[j]
            marklist[j]=t
    cities_dict=dict(marklist)
print(cities_dict)    
```

```python
cities_med_sorted = (list(cities_dict.values()))
cities_sorted = (list(cities_dict.keys()))
```
